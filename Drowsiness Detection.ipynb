# With Detection Line

```python
import cv2
import dlib
from scipy.spatial import distance
dic={}
for i in range(60):
    for j in range(14):
        dic[i+(j/14)]=1
def calculate_EAR(eye):
    A = distance.euclidean(eye[1], eye[5])
    B = distance.euclidean(eye[2], eye[4])
    C = distance.euclidean(eye[0], eye[3])
    ear_aspect_ratio = (A+B)/(2.0*C)
    return ear_aspect_ratio

cap = cv2.VideoCapture(0)
hog_face_detector = dlib.get_frontal_face_detector()
dlib_facelandmark = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
blink_count=0
frame_count=0
frame_threshold=0.22
EAR=0
count=0
s=0
i=0
j=0
blink_threshold=0.22

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces = hog_face_detector(gray)
    j+=1
    j%=14
    if j==0:
        i+=1
        i%=60
    for face in faces:

        face_landmarks = dlib_facelandmark(gray, face)
        leftEye = []
        rightEye = []

        for n in range(36,42):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            leftEye.append((x,y))
            next_point = n+1
            if n == 41:
                next_point = 36
            x2 = face_landmarks.part(next_point).x
            y2 = face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x2,y2),(0,255,0),1)

        for n in range(42,48):
            x = face_landmarks.part(n).x
            y = face_landmarks.part(n).y
            rightEye.append((x,y))
            next_point = n+1
            if n == 47:
                next_point = 42
            x2 = face_landmarks.part(next_point).x
            y2 = face_landmarks.part(next_point).y
            cv2.line(frame,(x,y),(x2,y2),(0,255,0),1)

        left_ear = calculate_EAR(leftEye)
        right_ear = calculate_EAR(rightEye)

        EAR = (left_ear+right_ear)/2
        EAR = round(EAR,2)
        prev=dic[i+(j/14)]
        dic[i+(j/14)]=EAR
        b=0
        if EAR<frame_threshold:
            frame_count+=1
        else:
            frame_count-=1
            frame_count=max(frame_count,0)
        if frame_count>=10 and b==0:
            blink_count=0
            cv2.putText(frame,"DROWSY",(20,100),
                cv2.FONT_HERSHEY_SIMPLEX,3,(0,0,255),4)
            cv2.putText(frame,"Sleepy Eyes",(20,400),
                cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),4)
            b=1
        blink_count=max(0,blink_count)
        if prev>blink_threshold and EAR<blink_threshold:
            blink_count+=1
        if prev<blink_threshold and EAR>blink_threshold:
            blink_count-=1
        if blink_count>=40 and b==0:
            
            cv2.putText(frame,"DROWSY",(20,100),
                cv2.FONT_HERSHEY_SIMPLEX,3,(0,0,255),4)
            cv2.putText(frame,"Too Much Blinking",(20,400),
                cv2.FONT_HERSHEY_SIMPLEX,2,(0,0,255),4)
            count+=1
            b=1
        if count==50:
            blink_count=0
            frame_count=0
            count=0
        cv2.putText(frame,str(blink_count),(20,100),
                cv2.FONT_HERSHEY_SIMPLEX,3,(0,0,255),4)
    cv2.imshow("Drowsiness Detection Test", frame)

    key = cv2.waitKey(1)
    if key == 27:
        break
cap.release()
cv2.destroyAllWindows()
```
